```python
def popularity(x):
    if x<=30:
        return "low"
    elif x<=70:
        return "Medium"
    else:
        return "High"
df['popularity_class']=df['track_popularity'].apply(popularity)
    
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['popularity_class']=le.fit_transform(df['popularity_class'])
```

```python
df.dtypes
```

```python
X=df[['track_duration_min','artist_followers','artist_popularity','explicit']]
Y=df['popularity_class']
```

```python
X
```

```python
Y
```

```python
df['explicit']=df['explicit'].astype(int)
```

```python
df['explicit'].dtypes
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
    
```

```python
from sklearn.metrics import accuracy_score, classification_report
y_predict=model.predict(x_test)
print(accuracy_score(y_test,y_predict))
print(classification_report(y_test,y_predict))
```

```python
#notice the medium value i.e. 1 that is giving the correct prediction because medium dominates the dataset 
#i.e.
#high=255 samples
#medium=606 samples
#low=184 samples
#accuracy is preety low i.e. 58 percent because poor features engineering as popularity must be predicted using data like energy, loudness, tempo, acousticness etc
# but we dont have that type of data due to which we end up with poor feature engineering, end up using duration_min, artist_popularity etc.
```

```python
new_popularity = pd.DataFrame([{
    'track_duration_min':3.02,
    'artist_followers':2882822,
    'artist_popularity':50,
    'explicit':1

}])
prediction = model.predict(new_popularity)
output=le.inverse_transform(prediction)
print(output)
```
